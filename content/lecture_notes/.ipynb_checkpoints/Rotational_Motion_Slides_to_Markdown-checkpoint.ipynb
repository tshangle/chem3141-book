## 2. Two model systems of angular motion in the X-Y plane: 2D rigid rotor and Particle on a Ring

- Two model systems of angular motion in the X-Y plane: 2D rigid rotor and Particle on a Ring
- 2D Rigid Rotor: Model for rotational motion of a molecule on a surface... Moment of inertia relates to atomic masses and bond lengths
- Particle on a ring: Model for electronic motion in a cyclic molecule… moment of inertia relates to electron mass and radius of the cyclic structure
